```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('tutorial2').getOrCreate()
```

```python
# load data
sdf = spark.read.csv('test1.csv', header=True, inferSchema=True)
sdf.show()
```

```python
# check the schema
sdf.printSchema()
```

```python
sdf.columns
```

```python
sdf.select('*')
```

```python
sdf.select(['Name', 'Age']).show()
```

```python
sdf.dtypes
```

```python
sdf.describe().show()
```

```python
#  create new column
from pyspark.sql.functions import col
sdf_2year = sdf.withColumn('Experience After 2 year', col('Experience')+2)
sdf_2year.show()
```

```python
# Drop the column
sdf = sdf_2year.drop('Experience After 2 year')
sdf.show()
```

```python
# Rename the columns
sdf = sdf.withColumnRenamed('Name', 'New Name')
sdf.show()
```
