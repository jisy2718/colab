```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('tutorial_04').getOrCreate()
```

```python
sdf = spark.read.csv('test1.csv', header=True, inferSchema=True)
```

```python
sdf.show()
```

### Filter Operations

```python
### Salary of the people
### 방법1
from pyspark.sql.functions import col
sdf.filter(col("Salary") <=35000).select(['Name','Age']).show()
```

```python
### 방법2
sdf.filter("Salary<=35000").select(['Name','Age']).show()
```

```python
### 방법3
sdf.filter((sdf['Salary']<=35000) & (sdf['Salary']>=30000)).show()
```

```python
### & 이용시 소괄호 쳐야함
sdf.filter((col("Salary") <=35000) & (col("Salary") >=30000)).show()
```

```python
### ~
sdf.filter(~(col("Salary") <=35000)).show()
```
