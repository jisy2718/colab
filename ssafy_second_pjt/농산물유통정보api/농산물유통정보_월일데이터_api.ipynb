````python
### json to pandas df
import pandas as pd
import json
from pandas import json_normalize
import os
import sys
import urllib.request
import pprint, json


months = ['m1', 'm2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'm10', 'm11', 'm12']
meta = []

category = []
mid_category = []
small_category = []
grade = []
per_volume = []
measure = []
year = []
month = []
price = []


# 검색 API 이용해서, 농산물 검색
p_yyyy = '2022'  # 시작년도
p_period='10'   # 시작년도 기준 과거로 n년까지 가져오기
p_returntype = 'json'
action='monthlySalesList'
p_convert_kg_yn='N'


for p_itemcategorycode, p_itemcode, p_kindcode, p_productrankcode, p_graderank in api_codes:
    # 파라미터 순서 일치해야함
    url = f"http://www.kamis.or.kr/service/price/xml.do?action={action}&p_yyyy={p_yyyy}&p_period={p_period}&p_itemcategorycode={p_itemcategorycode}&p_itemcode={p_itemcode}&p_kindcode={p_kindcode}&p_graderank={p_graderank}&p_convert_kg_yn={p_convert_kg_yn}&p_cert_key={p_cert_key}&p_cert_id={p_cert_id}&p_returntype={p_returntype}"
   
    # API요청
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        tmp = json.loads(response_body.decode('utf-8'))
    else:
        print("Error Code:" + rescode)
    # DF로 변환하기
    print(tmp['condition'])
    print(tmp['error_code'])
    print(tmp.keys())
    print('-'*100)
    # 가격정보 존재하면 진행
    if 'price' in tmp:
        prices = tmp['price']
    # 1. 소매인 경우
    ## 1-1 도/소매 같이 나온 경우
    if type(prices)==list:
    
        for data in prices:
            if data['productclscode'] == '01':  # 소매
                # 2. tmp_df
                tmp_df = json_normalize(data['item'])
         
                
                # 데이터 개수
                years = list(tmp_df.loc[:,'yyyy'])
                years.sort()
                # print(years)
                # 1. 카테고리, 품목, 상품, 등급, 단위용량, 단위명
                meta.append(data['caption'].split('>'))
                cate =  data['caption'].split('>')[1].strip()
                mid_cate =  data['caption'].split('>')[2].strip()
                small_cate =  data['caption'].split('>')[3].strip()
                grad  = data['caption'].split('>')[4].strip()
                volume =  data['caption'].split('>')[5].strip()
                
                # volume은 20kg 과 같이 되어있는데, 이를 20, kg 으로 분리하기
                per_v = ''
                u = ''
                for i in range(len(volume)):
                    cur = volume[i]
                    if ord(cur) in set(range(48,58)): # '0' ~ '9'
                        per_v += cur
                    else:
                        u += cur
                category.extend(         [cate]*(len(years)*12)        )
                mid_category.extend(     [mid_cate]*(len(years)*12)    )
                small_category.extend(   [small_cate]*(len(years)*12)  )
                grade.extend(            [grad]*(len(years)*12)        )
                per_volume.extend(       [per_v]*(len(years)*12)       )
                measure.extend(             [u]*(len(years)*12)           )
                
                ## 2-1 year 가져오기
                for yyyy in years:
                    year.extend([yyyy]*12)
                
                ## 2-2 month 가져오기
                tmp = ['01','02','03','04','05','06','07','08','09','10','11','12']*len(years)
                for i in range(len(years)):
                    yyyy = years[i]
                    for j in range(i*12, i*12 + 12):
                        mm = tmp[j]
                        yyyy_mm_dd = str(yyyy) +'-' + mm + '-' +'01'
                        tmp[j] = yyyy_mm_dd
                month.extend(tmp)
                ## 2-3 price 가져오기
                for yyyy in range(len(years)):
                    price.extend(list(tmp_df.loc[yyyy,months]))
                    
    ## 1-2 1개만 나온 경우
    else:
        data = prices
        if data['productclscode'] == '01':  # 소매
            # 2. tmp_df
            tmp_df = json_normalize(data['item'])
            
            # 데이터 개수
            years = list(tmp_df.loc[:,'yyyy'])
            years.sort()
            # print(years)
            # 1. 카테고리, 품목, 상품, 등급, 단위용량, 단위명
            meta.append(data['caption'].split('>'))
            cate =  data['caption'].split('>')[1].strip()
            mid_cate =  data['caption'].split('>')[2].strip()
            small_cate =  data['caption'].split('>')[3].strip()
            grad  = data['caption'].split('>')[4].strip()
            volume =  data['caption'].split('>')[5].strip()
            
            # volume은 20kg 과 같이 되어있는데, 이를 20, kg 으로 분리하기
            per_v = ''
            u = ''
            for i in range(len(volume)):
                cur = volume[i]
                if ord(cur) in set(range(48,58)): # '0' ~ '9'
                    per_v += cur
                else:
                    u += cur
            category.extend(         [cate]*(len(years)*12)        )
            mid_category.extend(     [mid_cate]*(len(years)*12)    )
            small_category.extend(   [small_cate]*(len(years)*12)  )
            grade.extend(            [grad]*(len(years)*12)        )
            per_volume.extend(       [per_v]*(len(years)*12)       )
            measure.extend(             [u]*(len(years)*12)           )
            
            ## 2-1 year 가져오기
            for yyyy in years:
                year.extend([yyyy]*12)
            print(year)
            ## 2-2 month 가져오기
            tmp = ['01','02','03','04','05','06','07','08','09','10','11','12']*len(years)
            for i in range(len(years)):
                yyyy = years[i]
                for j in range(i*12, i*12 + 12):
                    mm = tmp[j]
                    yyyy_mm_dd = str(yyyy) +'-' + mm + '-' +'01'
                    tmp[j] = yyyy_mm_dd
            month.extend(tmp)
            ## 2-3 price 가져오기
            for yyyy in range(len(years)):
                price.extend(list(tmp_df.loc[yyyy,months]))
 
````
